```python
import numpy as np
import matplotlib.pyplot as plt

fd = 1                                  # Frequência do desvio
t = np.linspace(0,30,2000)
signal = np.sin(2*np.pi*2.0*t)          # Onda portadora
data=(np.random.rand(10)>0.5)
data_sequence = np.repeat(data,200)     # Sinal Digital


fsk = np.zeros(2000)                    # Sinal FSK

for i, bit in enumerate(data):
  start = i * 200
  end = (i + 1) * 200
  if bit == True:
    fsk[start:end] = np.sin(2*np.pi*(2.0 - fd)*t[start:end])
  else:
    fsk[start:end] = np.sin(2*np.pi*(2.0 + fd)*t[start:end])

plt.figure(1)
plt.plot(t, signal)
plt.figure(2)
plt.plot(t,data_sequence, linestyle='--', color='r', linewidth=4)
plt.plot(t,fsk)
plt.grid(True)
plt.show()
```
