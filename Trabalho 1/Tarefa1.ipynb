```python
# Exercicio 1: Gerador Congruente Linear Misto
import matplotlib.pyplot as plt
a = 5                     # Constante multiplicadora
c = 3                     # Incremento
m = 16                    # M√≥dulo
xo = 7                    # Semente
amostras = 40             # Quantidade de n√∫meros gerados
numeros_gerados = [xo]    # Lista para armazenar as amostras geradas.
cont = 0

 # Executa um loop tantas vezes quanto o n√∫mero de amostras
 # para calcular os n√∫meros gerados.
for i in range(amostras):
  if i == 0:
    x = xo
  x = (a*x + c)%(m)       # F√≥rmula do Gerador Congruente Linear Misto
  numeros_gerados.append(x)

# a) O per√≠odo desse gerador.
# Conta quantas itera√ß√µes s√£o necess√°rias at√© que o n√∫mero gerado se repita
# (volte √† semente xo). Esse valor corresponde ao per√≠odo do gerador.

for n in numeros_gerados[1:]:
  cont += 1
  if n == xo:
    break
T = cont               # Per√≠odo do gerador
print(numeros_gerados)
plt.bar(range(len(numeros_gerados)),numeros_gerados)
print('Per√≠odo:',T)
```

```python
# Exercicio 2: Gerador de Variaveis Aleatoria com Distribuicao de Poisson
import math
import numpy as np
import matplotlib.pyplot as plt

media = 3                      # Numero medio de chegadas num intervalo de tempo
N = 100000                      # N√∫mero de amostras a serem geradas
u = np.random.uniform(0, 1, N) # Gera N amostras aleat√≥rias uniformes u ~ U(0,1)
poisson = []

# Converte cada valor uniforme u em uma vari√°vel Poisson
# ao encontrar o menor x tal que CDF(x) > u

for element in u:
  i = 0
  pr = math.exp(-media)
  CDF = pr
  while element >= CDF:
    pr = media/(i+1)*pr        # Probabilidade de forma recursiva
    CDF = CDF + pr
    i += 1
  poisson.append(i)
print(poisson)
plt.hist(poisson, bins = np.arange(0,max(poisson)+2)-0.5, edgecolor ='black', density = True)
plt.show()

# a) A probabilidade de que exatamente 5 chamadas sejam recebidas em um minuto espec√≠fico
x = 5
poisson = np.array(poisson)   # Converte a lista em um array Numpy.
mask = poisson == x           # M√°scara booleana: True onde os elementos do array poisson sao iguais ao x.
quantidade = np.sum(mask)     # Soma os valores True da m√°scara booleana
# Divide o n√∫mero de amostras igual a x pelo n√∫mero total de amostras N para calcular Pr(X = x).
Pr = quantidade/N
print('A probabilidade de que exatamente 5 chamadas sejam recebidas em um minuto espec√≠fico:', Pr)

# b) A probabilidade de que no m√°ximo 2 chamadas sejam recebidas em um minuto espec√≠fico
y = 2
mask1 = poisson <= y          # M√°scara booleana: True onde os elementos do array poisson sao menores ou iguais ao y.
quantidade1 = np.sum(mask1)   # Soma os valores True da m√°scara booleana
# Divide o n√∫mero de amostras menores ou iguais ao y pelo n√∫mero total de amostras N para calcular Pr(Y <= y).
Pr1 = quantidade1/N
print('A probabilidade de que no m√°ximo 2 chamadas sejam recebidas em um minuto espec√≠fico:', Pr1)
```

```python
# Exercicio 3: Gerador de Variaveis Aleatoria com Distribuicao Binomial
import numpy as np
import math
import matplotlib.pyplot as plt
q = 0.25                           # Probabilidade de sucesso
n = 10                             # N√∫mero de experimentos
N = 100000                         # N√∫mero de amostras a serem geradas
u = np.random.uniform(0, 1, N)     # Gera N amostras aleat√≥rias uniformes u ~ U(0,1)
binomial = []

# Converte cada valor uniforme u em uma vari√°vel Binomial ao encontrar o menor x
# tal que CDF(x) > u

for element in u:
  i = 0
  pr = (1-q)**(n)
  CDF = pr
  while element >= CDF and i < n:
      pr = (n-i)/(i+1)*(q/(1-q))*pr  # Probabilidade de forma recursiva
      CDF = CDF + pr
      i += 1
  binomial.append(i)

print(binomial)
plt.hist(binomial, bins = np.arange(0,max(binomial)+2)-0.5, edgecolor ='black', density = True)
plt.show()

# a) A probabilidade de o aluno acertar exatamente 3 quest√µes
x = 3
# Converte a lista em um array Numpy.
binomial = np.array(binomial)
# M√°scara booleana: True onde os elementos do array binomial sao iguais ao x.
mask = binomial == x
# Soma os valores True da m√°scara booleana
quantidade = np.sum(mask)
# Divide o n√∫mero de amostras igual a x pelo n√∫mero total de amostras N para
# calcular Pr(X = x).
Pr = quantidade/N
print('A  probabilidade de o aluno acertar exatamente 3 quest√µes:', Pr)

# b) A probabilidade de ele acertar no m√°ximo 2 quest√µes
y = 2
# M√°scara booleana: True onde os elementos do array binomial sao menores ou iguais ao y.
mask1 = binomial <= y
# Soma os valores True da m√°scara booleana
quantidade1 = np.sum(mask1)
# Divide o n√∫mero de amostras menores ou iguais ao y pelo n√∫mero total de
# amostras N para calcular Pr(Y <= y).
Pr1 = quantidade1/N
print('A probabilidade de ele acertar no m√°ximo 2 quest√µes:', Pr1)

# c) A m√©dia e o desvio padr√£o
# M√©dia da variable aleatoria X
media = np.sum(binomial)/N
print('A M√©dia:', media)
media_analitica = n*q
print('A M√©dia anal√≠tica:', media_analitica)
# Vari√¢ncia
variance = np.sum((binomial - media)**2)/(N-1);
variance_analitica = n*q*(1-q)
# Desvio padr√£o
desvio_padrao = math.sqrt(variance)
desvio_padrao_analitico = math.sqrt(variance_analitica)
print('O desvio padr√£o da amostra:',desvio_padrao)
print('O desvio padr√£o anal√≠tico:',desvio_padrao_analitico)
```

```python
# Exercicio 4: Gerador de Variaveis Aleatoria com Distribuicao de Poisson
import math
import numpy as np
import matplotlib.pyplot as plt

media = 3                      # Numero medio de chegadas num intervalo de tempo
N = 100000                     # N√∫mero de amostras a serem geradas
u = np.random.uniform(0, 1, N) # Gera N amostras aleat√≥rias uniformemente distribu√≠dos u ~ U(0,1)
poisson = []

# Converte cada valor uniforme u em uma vari√°vel Poisson ao encontrar
# o menor x tal que CDF(x) > u

for element in u:
  i = 0
  pr = math.exp(-media)
  CDF = pr
  while element >= CDF:
    pr = media/(i+1)*pr               # Probabilidade de forma recursiva
    CDF = CDF + pr
    i += 1
  poisson.append(i)
print(poisson)

# A probabilidade de que haver√° ao menos 3 falhas durante uma semana espec√≠fica
x = 2
# Converte a lista em um array Numpy.
poisson = np.array(poisson)
# M√°scara booleana: True onde os elementos do array poisson sao menores ou iguais ao x.
mask = poisson <= x
# Soma os valores True da m√°scara booleana
quantidade = np.sum(mask)
# Divide o n√∫mero de amostras menores ou iguais a x pelo n√∫mero total de amostras N para calcular Pr(X <= x).
p = quantidade/N
 # Pr(X > x) = 1 - P(X <= x-1)
q = 1 - p
print('A probabilidade de que haver√° ao menos 3 falhas durante uma semana espec√≠fica:', q)

# Histograma
plt.hist(poisson, bins=np.arange(0, max(poisson)+2)-0.5, edgecolor='black', density=True)
plt.show()
```

```python
# Exercicio 5: : Gerador de Variaveis Aleatoria com Distribuicao Exponencial
import numpy as np
import matplotlib.pyplot as plt
import math

beta = 2                     # Tempo m√©dio entre eventos (m√©dia da distribui√ß√£o)
Lambda = 1/beta              # Numero medio de chegadas num intervalo de tempo
N = 100000                   # N√∫mero de amostras a serem geradas
u = np.random.uniform(0, 1, N)  # Gera N amostras aleat√≥rias uniformemente distribu√≠das u~U(0,1)
exponencial = np.array([])

# M√©todo da transformada inversa: x = F^(-1)(u)

for element in u:
  x = -math.log(element)/Lambda
  exponencial = np.append(exponencial, x)

# Curva te√≥rica da distribui√ß√£o exponencial

# Gera uma matriz com elementos igualmente espa√ßados de 0 at√© 10 (passo 0.1)
y = np.arange(0, 10, 0.1)
# Fun√ß√£o de densidade de probabilidade (pdf) de uma distribui√ß√£o exponencial
pdf = Lambda*(np.exp(-Lambda*y))

# Gr√°fico
plt.plot(y, pdf)
plt.hist(exponencial, bins=100, edgecolor='black', density=True)
plt.show()

# b) A probabilidade de que o tempo de espera at√© a chegada do pr√≥ximo
#    cliente seja inferior a 1 minuto
valor = 1
# M√°scara booleana: True onde os elementos do array exponencial sao menores ou iguais ao valor.
mask = exponencial <= valor
# Soma os valores True da m√°scara booleana
quantidade = np.sum(mask)
# Divide o n√∫mero de amostras menores ou iguales a valor pelo n√∫mero total de
# amostras N para calcular P(X <= valor).
Pr = quantidade/N
print('A probabilidade de que o tempo de espera at√© a chegada do pr√≥ximo cliente seja inferior a 1 minuto:', Pr)

# c) A probabilidade de que o tempo de espera at√© a chegada do pr√≥ximo cliente seja superior a 4 minutos
valor1 = 4
# M√°scara booleana: True onde os elementos do array exponencial sao maiores ou iguais ao valor1.
mask1 = exponencial >= valor1
# Soma os valores True da m√°scara booleana
quantidade1 = np.sum(mask1)
# Divide o n√∫mero de amostras maiores ou iguales a valor1 pelo n√∫mero total de amostras N para calcular P(X >= valor1).
Pr1 = quantidade1/N
print('A probabilidade de que o tempo de espera at√© a chegada do pr√≥ximo cliente seja superior a 4 minutos:', Pr1)
```

```python
# Exercicio 6:  Gerador de Variaveis Aleatoria com Distribuicao Geom√©trica
import numpy as np
import math
import matplotlib.pyplot as plt

N = 100000                                # N√∫mero de amostras a serem geradas
p = 1/6                                   # Probabilidade de succeso
u = np.random.uniform(0, 1, N)            # Gera N amostras aleat√≥rias uniformes u ~ U(0,1)
dist_geometrica = []

# Converte cada valor uniforme u em uma vari√°vel com Distribuicao Geom√©trica
# ao encontrar o menor x tal que CDF(x) > u

for element in u:
  i = 1
  Pr = p
  CDF = Pr
  while CDF <= element:
    Pr = (1-p)*Pr                        # Probabilidade de forma recursiva
    CDF = CDF + Pr
    i += 1
  dist_geometrica.append(i)

print(dist_geometrica)
plt.hist(dist_geometrica, bins = np.arange(1, max(dist_geometrica) + 2)-0.5, edgecolor ='black', density=True)
plt.show()

# a) A probabilidade de que o jogador ganhe o jogo exatamente no terceiro lan√ßamento
cont = 0
for element in dist_geometrica:
  if element == 3:
    cont += 1
pr = cont/N
print('A probabilidade de que o jogador ganhe o jogo exatamente no terceiro lan√ßamento:',pr)

# b) A probabilidade de que ele precise lan√ßar o dado pelo menos 4 vezes para ganhar o jogo
cont1 = 0
for element in dist_geometrica:
  if element >= 4:
    cont1 += 1
pr1 = cont1/N
print('A probabilidade de que ele precise lan√ßar o dado pelo menos 4 vezes para ganhar o jogo:',pr1)

# c) A m√©dia e o desvio padr√£o de X
media = sum(dist_geometrica)/N
print('A M√©dia:', media)
media_analitica = 1/p
print('A M√©dia anal√≠tica:', media_analitica)

numerador = 0
for element in dist_geometrica:
  numerador = numerador + (element - media)**2
variance = numerador/(N-1)
variance_analitica = (1-p)/(p**2)
desvio_padrao = math.sqrt(variance)
desvio_padrao_analitico = math.sqrt(variance_analitica)
print('O desvio padr√£o da amostra:', desvio_padrao)
print('O desvio padr√£o anal√≠tico:', desvio_padrao_analitico)
```

```python
# Exercicio 7: M√©todo da inversa para gerar amostras da distribui√ß√£o f(x) = 3x^2
import numpy as np
import matplotlib.pyplot as plt

N = 1000000                            # N√∫mero de amostras a serem geradas
u = np.random.uniform(0, 1, N)         # Gera N amostras aleat√≥rias uniformemente distribu√≠dos entre 0 e 1
x = np.pow(u, 1/3)                     # Inversa

# Curva te√≥rica
y = np.linspace(0, 1, 1000)
PDF = 3*(y**2)

plt.plot(y, PDF)
plt.hist(x, bins = 100,edgecolor='black', density=True)
plt.show()
```

```python
# Exercicio 8: M√©todo da aceita√ß√£o-rejei√ß√£o para gerar amostras da distribui√ß√£o f(x) = 3x^2
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

N = 1000000  # N√∫mero de amostras a serem geradas
def f(x):
    return 3 * (x**2)  # PDF objetivo

# a) Verificar se a fun√ß√£o de densidade de probabilidade f(x) = 3x^2 √© v√°lida:
#    f(x) >= 0 e a √°rea sob a curva == 1

# Integral de f(x)
# Gerar u n√∫meros aleat√≥rios uniformes u ~ U(0,1)
# Para cada n√∫mero gerado, calcular f(Ui)
# A integral e igual a m√©dia de todos os valores f(Ui)

u = np.random.uniform(0, 1, N)
area = np.sum(f(u))/N

# Para que uma fun√ß√£o de densidade seja v√°lida, a fun√ß√£o deve ser n√£o negativa
# e a √°rea sob a curva dessa fun√ß√£o deve ser 1.

x = np.linspace(0, 1, 100)

if np.all(f(x) >= 0) and np.isclose(area,1,rtol=1e-2):
    mensagem = 'Fun√ß√£o de densidade v√°lida'
else:
    mensagem = 'Fun√ß√£o de densidade nao v√°lida'
print(mensagem)

# b) Encontrar uma constante c adequada para aceita√ß√£o-rejei√ß√£o

def g(x):
    return 1  # PDF proposta g(x) uniforme

c = max(f(x) / g(x))  # max(f(x)/g(x))
print('Constante c adequada:', c)

# c) Gera√ß√£o das amostras
Y = np.random.uniform(0, 1, N)  # Amostras Y ~ g(x)
U = np.random.uniform(0, 1, N)  # U ~ Uniforme(0,1)

amostras_aceitas = []
for i in range(N):
    if U[i] <= f(Y[i]) / (c * g(Y[i])):
        amostras_aceitas.append(Y[i])

# Gr√°ficos
plt.plot(x, f(x))                                                           # PDF anal√≠tica
plt.hist(amostras_aceitas, bins = 100, edgecolor = 'black', density=True)   # Histograma normalizado
plt.show()
```
